```python
from math import isfinite, sqrt
import numpy as np
import matplotlib.pyplot as plt
```

```python
def brownian_motion(x0: float, size: int, dt: float, delta: float):
    if (not isfinite(x0) or size < 0 or not isfinite(dt) or dt <= 0.0
            or not isfinite(delta) or delta <= 0.0):
        raise ValueError

    scale = delta * sqrt(dt)
    r = np.random.normal(scale=scale, size=size)
    return np.cumsum(r) + x0
```

```python
delta = 2
T = 20.0
N = 2000
dt = T / N
x0 = 50
x = brownian_motion(x0=x0, size=N, dt=dt, delta=delta)

t = np.linspace(0.0, N * dt, N)
```

```python
fig = plt.figure(figsize=(16,9), dpi=300)
subplot = fig.add_subplot(111, facecolor = '#FFFFFF')
subplot.plot(t, x, color = 'red', lw = 2, label = 'x')
plt.legend()
plt.show()
```

# Фильтрация

```python
from scipy import fft
from scipy.fftpack import fftshift
from scipy.signal.windows import gaussian
import matplotlib.pyplot as plt

sigma = 50
window = gaussian(50, std=sigma)
plt.plot(window)
plt.title(f'Gaussian window ($\sigma$={sigma})')
plt.ylabel("Amplitude")
plt.xlabel("Sample")

plt.figure()
A = fft(window, 2048) / (len(window) / 2.0)
freq = np.linspace(-0.5, 0.5, len(A))
response = 20 * np.log10(np.abs(fftshift(A / abs(A).max())))
plt.plot(freq, response)
plt.axis([-0.5, 0.5, -120, 0])
plt.title(f'Frequency response of the Gaussian window ($\sigma$={sigma})')
plt.ylabel('Normalized magnitude [dB]')
plt.xlabel('Normalized frequency [cycles per sample]')
```

```python
from scipy.signal import lfilter

def windows_filter(b, x):
    a = np.zeros(len(b))
    a[0] = 1
    filter_x = lfilter(b, a, x)
    filter_x[0:len(b)] = filter_x[len(b)]
    return filter_x * (np.sum(x) / np.sum(filter_x))
```

```python
fig = plt.figure(figsize=(16,9), dpi=300)
subplot = fig.add_subplot(111, facecolor = '#FFFFFF')
subplot.plot(t, x, color = 'red', lw = 2, label = 'x')
# subplot.plot(t, windows_filter(window, x), color = 'blue', lw = 2, label = 'filter_x')
subplot.plot(t, windows_filter(gaussian(50, std=10), x), color = 'blue', lw = 2, label = 'filter_x')
subplot.plot(t, windows_filter(gaussian(50, std=100), x), color = 'green', lw = 2, label = 'filter_x')
subplot.plot(t, windows_filter(gaussian(10, std=50), x), color = 'brown', lw = 2, label = 'filter_x')
plt.legend()
plt.show()
```
